**Objetivo General**

Analizar el comportamiento de los usuarios, sus preferencias de contenido y los problemas de soporte técnico para descubrir qué factores impulsan el consumo y detectar fallas críticas en el servicio.   

1. **Conclusiones**

- **Evidencia**: Los usuarios Premium pasan casi el doble de tiempo viendo contenido que los del plan Básico (~1,081 minutos al mes frente a ~557) y eligen principalmente Drama y Documentales. Además, la edad no influye en lo que eligen ver (la correlación dio 0.00), y los clientes Premium de Chile y Perú son los que más reclamos técnicos hacen.

- **Interpretación**: Esto nos indica que pagar un plan más caro hace que la gente consuma mucho más y elija series más largas. Por otro lado, los gustos son iguales para jóvenes y grandes (la edad no define el consumo). Finalmente, los problemas técnicos no ocurren por usar mucho la app, sino por fallas específicas en esos dos países.

- **Conclusión**: Se concluye que el negocio debe enfocar sus recomendaciones de catálogo según el tipo de plan del usuario y dejar de lado los rangos de edad, ya que los gustos son transversales. Asimismo, se detecta una alerta comercial crítica: la empresa debe revisar de forma urgente el servicio técnico Premium en Chile y Perú para solucionar los problemas antes de perder a los clientes que más pagan.
